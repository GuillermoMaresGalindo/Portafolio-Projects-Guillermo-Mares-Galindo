# Data Sampling Methods in Python


Data Sampling forms the essential part of the majority of research, scientific and data experiments. It is one of the most important factors which determines the accuracy of your research or survey result.

If sampling is done wrong, it will lead to biases that affect the accuracy of your research/survey results. To avoid selection biases, we have to carefully choose a subset of a population that can be representative of the group as a whole.

**Population** refers to the complete collection of observations we want to study, and a **sample** is a subset of the target population.

### **Probability vs Non-Probability Sampling**

#### **1. Simple Random Sampling**
<small>Each person in a population is given an identification, such as a number, and those chosen to be included in the sample are chosen at random, often using an automated software program. This is the simplest type of probability sampling, known as a simple random sample.</small>

**When to Use**: <small>Use simple random sampling when the population is homogeneous, meaning there is little variation in the characteristics being studied, and you want each individual to have an equal chance of selection. It’s ideal when the population size is manageable and a complete list of the population is available.</small>

#### **2. Stratified Random Sampling**
<small>A simple random sample is the simpler version compared to a stratified random sample. In this approach, the population is broken down into sub-groups, such as men and women, and a simple random sample is collected from within each sub-group. This provides a random sample that is representative of a broader population and its particular make-up, such as the population of a nation.</small>

**When to Use**: <small>Use stratified random sampling when the population is heterogeneous and you want to ensure representation from all sub-groups (strata) within the population. It’s especially useful when specific sub-groups are important to the research outcome, such as age groups, income levels, or educational backgrounds.</small>

#### **3. Cluster Sampling**
<small>A population is separated into clusters in cluster sampling that are distinct but reflect a variety of groups; cities are often used as clusters. A few clusters are randomly chosen to participate in the research from the list of clusters.</small>

**When to Use**: <small>Use cluster sampling when it’s difficult or expensive to collect data from a large population spread out geographically. It’s also suitable when it’s more feasible to collect data from groups (clusters) rather than individuals. This method is often used in large-scale surveys, like those conducted at the national level.</small>

#### **4. Systematic Sampling**
<small>In systematic sampling, you select participants from a list at regular intervals, such as every 5th person on the list. This method is straightforward and ensures a spread of participants across the population.</small>

**When to Use**: <small>Use systematic sampling when you have a list of the entire population and want a simple and quick way to select a sample. It's effective when the population is evenly distributed and doesn't have a particular order that could introduce bias.</small>

### **Different Non-Probability Sampling Methods**

#### **1. Convenience Sampling**
<small>This type of sampling, as its name suggests, depends on having quick access to research participants. These may be friends, individuals passing by on the street, or students taking a course at a university. While convenience sampling is fast and simple, it does not provide data that can be generalized to a larger population.</small>

#### **2. Snowball Sampling**
<small>Snowball sampling operates by enlisting a few sample participants, who then enlist acquaintances of their own to join the sample. This type of sampling is effective for reaching highly niche people who are likely to know others who fulfill the requirements.</small>

#### **3. Quota Sampling**
<small>In this type of sampling, a population is segmented into subgroups based on factors like geography or age, and goals are established for the number of responses required from each category. The primary distinction between quota sampling and stratified random sampling is that quota sampling does not use a random sampling approach.</small>

<small>For instance, a researcher could use a convenience sample with predetermined quotas to guarantee that there are equal numbers of men and women included, but this approach would still not give every member of the population a fair chance of being chosen, making it clear that it is not a probability sample.</small>

#### **4. Purposive or Judgmental Sampling**
<small>Using this type of sampling method, the researcher chooses the sample based on their understanding of who would meet the study’s eligibility requirements.</small>

<small>For instance, a purposeful sample can consist only of Ph.D. candidates in a certain field. This selection approach may be used to investigate certain traits. Bias may, however, be introduced since the researcher may have an impact on the people who are chosen to participate in the study.</small>

## Random sampling

```python
import pandas as pd
import numpy as np
```

This is a demographic survey conducted by the market research department. The focus of this dataset is to have the customers' preferences regarding the computer brand.

data link:
https://www.kaggle.com/datasets/samanemami/market-research-survey

```python
data=pd.read_csv("data/CompleteResponses_full.csv")
```

```python
data.head()
```

Each person in a population is given an identification, such as a number, and those chosen to be included in the sample are chosen at random, often using an automated software program. This is the simplest type of probability sampling, known as a simple random sample.

```python
# 1. Simple Random Sampling
# Sample 2 random rows from the DataFrame
simple_random_sample = data.sample(n=2, random_state=1)
print("\nSimple Random Sampling:")
display(simple_random_sample)
```

## Stratified Random Sampling

```python
# 2.1 Stratified Random Sampling
# We will stratify by 'elevel' column, select a single row from each group
stratified_sample = data.groupby('elevel').apply(lambda x: x.sample(n=1, random_state=1)).reset_index(drop=True)
print("\nStratified Random Sampling:")
display(stratified_sample)
```

```python
# 2.2 Stratified Random Sampling
# We will stratify by 'elevel' column,, select 10% of rows from each group
stratified_sample = data.groupby('elevel').sample(frac=0.1, random_state=1)
print("\nStratified Random Sampling:")
display(stratified_sample)
```

```python
# 2.3 Stratified Random Sampling
# We will stratify by 'elevel' column, select 2 rows from each group
stratified_sample = data.groupby('elevel').sample(n=2, random_state=1)
print("\nStratified Random Sampling:")
display(stratified_sample)
```

## Cluster Sampling

```python
# 3. Cluster Sampling
# Treat 'zipcode' as clusters, randomly select 2 clusters, then sample all members of these clusters
clusters = data['zipcode'].unique()
selected_clusters = np.random.choice(clusters, size=2, replace=False)
cluster_sample = data[data['zipcode'].isin(selected_clusters)]
print("\nCluster Sampling:")
display(cluster_sample)
```

## Systematic Sampling

Systematic sampling or interval sampling is a probability sampling technique used to select a sample from a population in a systematic way. In systematic sampling, the population is first ordered in some way, and then every nth individual is selected for inclusion in the sample, where “n” is a predetermined sampling interval. 

```python
# 4.1 Systematic Sampling
# Sample every 2nd row
systematic_sample = data.iloc[::2]
print("\nSystematic Sampling every n rows:")
display(len(systematic_sample))
display(systematic_sample.head())
```

```python
# 4.2 Systematic Sampling
# Sample with a sample size of 5
sample_size=5
pop_size = len(data)
display(pop_size)
interval=pop_size//(sample_size-1)
systematic_sample = data.iloc[::interval]
print("\nSystematic Sampling with range size:")
display(systematic_sample)
```
